# **НЭ. Задание `A3`**

[SmartLMS](https://edu.hse.ru/mod/quiz/view.php?id=1618961)

**Необходимо выбрать корректную статистику для описания определенного признака**. Необходимо понимание типов признаков и статистик (МЦТ и меры вариантивности) - посмотреть можно в [Хэндбуке, главе #6](https://edu.hse.ru/course/view.php?id=236211&section=6)  

## **Вопрос 1**

>*В наборе данных информация об уровне владения иностранным языком студентов первого курса закодирована по системе CEFR (A1, A2, B1, B2, C1 или С2). Какую меру вариативности корректно использоваться для данного признака?*
>- *Мода*
>- *Количество уникальных значений*
>- *Медиана*
>- *Стандартное отклонение*

**Ответ**: Количество уникальных значений

---

## **Вопрос 2**

>*Какую меру измерения центральной тенденции корректно использовать при работе с категориальной бинарной переменной, которая отвечает за пол респондента и закодирована числами 0 и 1?*
>- *Медиана*
>- *Размах*
>- *Среднее арифметическое*
>- *Интерквартильное расстояние*

**Ответ**: Среднее арифметическое

---

## **Вопрос 3**

>*Какую меру измерения центральной тенденции корректно использовать при работе с номинальной переменной, которая отвечает за профессию респондента?*
>- *Мода*
>- *Среднее арифметическое*
>- *Количество уникальных значений*
>- *Дисперсия*

**Ответ**: Мода

>*Какую описательную статистику некорректно использовать при работе с количественной непрерывной переменной?*

## **Вопрос 5**

>*Какую меру измерения центральной тенденции можно использовать при работе с категориальными переменными, которые не закодированы числами?*

## **Вопрос 6**

>*Какая мера разброса данных менее всего чувствительна к выбросам?*
>- *Среднеквадратичное отклонение*
>- *Дисперсия*
>- *Межквартильный размах / интерквартильный размах*
