## **Вопрос 1**

Вам предстоит работать с датасетом, который содержит информацию о продаже автомобилей в Армении `train1_C.csv`:


| **Признак**  | **Описание**                                |
|--------------|---------------------------------------------|
| Car Name     | марка и модель автомобиля                   |
| Year         | год выпуска автомобиля                       |
| Region       | локация, где автомобиль был зарегистрирован |
| FuelType     | тип двигателя автомобиля                     |
| Mileage      | пробег автомобиля (км)                       |
| Price        | цена автомобиля (драмы)                      |
| Owners       | количество предыдущих владельцев            |


>*Постройте корреляционную матрицу. Найдите переменную, которая имеет самую сильную корреляцию с переменной `Year` (не включая саму `Year`). Признак `coded` учитывать не нужно. В ответ выпишите модуль полученного значения корреляции. Ответ округлите до двух знаков.*

```python
df = pd.read_csv('data/train1_C.csv')
```

`Year` - числовая переменная. При поиске самой сильной корреляции следует смотреть как на корр Писона (с числовыми признаками), так и Пирсона (категориальные признаки)

```python
# df.corr() -- матрица корреляции (датафрейм)
# .sort_values() - сортируем по значению корреляции;
# key=abs - сортируем по МОДУЛЮ корреляции (слабые сверху, сильные снизу, dне зависимости от знака)

df.corr(numeric_only=True)['Year'].sort_values(key=abs)
```

```python
df.corr(numeric_only=True, method='spearman')['Year'].sort_values(key=abs)
```

Самая сильная корреляция с `Mileage` - это числовой признак, поэтому вписываем корреляцию Пирсона (просят модуль)

**Ответ**: 0.64
