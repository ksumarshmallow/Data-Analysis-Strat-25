## **Вопрос 1**

>*Экономист Дарья занимается исследованием рынка квартир. Определите тип переменных, с которыми она работает.*

Варианты для выбора:
- Номинальная
- Ранговая
- Бинарная
- Дискретная
- Непрерывная

**Ответ**: 

| **Признак**                                | **Тип переменной** |
| ------------------------------------------ | ------------------ |
| Цвет дома                                  | Номинальная        |
| Класс квартиры (от 1 до 5)                 | Ранговая           |
| Факт наличия станции метро в радиусе 700 м | Бинарная           |
| Количество комнат в квартире               | Дискретная         |
| Цена квартиры                              | Непрерывная        |

## **Вопрос 4**

>*Сопоставьте каждую переменную с типом, к которому она относится (трассы).*

Варианты для выбора:
- Категориальная номинальная
- Количественная непрерывная
- Категориальная бинарная
- Категориальная порядковая
- Количественная дискретная

**Ответ**: 

| **Признак**                                         | **Тип переменной**         |
| --------------------------------------------------- | -------------------------- |
| Часть света (например, Австралия, Европа)           | Категориальная номинальная |

| Есть ли на трассе опасные участки                   | Категориальная бинарная    |
| Уровень сложности трассы (низкий, средний, высокий) | Категориальная порядковая  |
| Количество площадок для отдыха                      | Количественная дискретная  |

## **Вопрос 5**

>*Аналитик Евгения исследует модели автомобилей. Определите тип переменных.*

Варианты для выбора:
- Категориальная бинарная
- Категориальная номинальная
- Количественная дискретная
- Категориальная порядковая
- Количественная непрерывная

**Ответ**: 

| **Признак**                                      | **Тип переменной**         |
| ------------------------------------------------ | -------------------------- |
| Наличие дизельного двигателя                     | Категориальная бинарная    |
| Марка автомобиля                                 | Категориальная номинальная |
| Количество мест в автомобиле                     | Количественная дискретная  |
