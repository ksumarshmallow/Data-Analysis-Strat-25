В среднем, на одном федеральном канале за день выходит 20-30 телепрограмм. Мы проанализируем отчет [Mediascope](https://www.google.com/url?q=https%3A%2F%2Fmediascope.net%2Fservices%2Fmedia%2Fmedia-audience%2Ftv%2Fnational-and-regional%2Fdescription%2F) о программах-лидерах c точки зрения телевизионной аудитории за первую неделю сентября.


| Признак             | Описание                                |
| ------------------- | --------------------------------------- |
| Дата |  дата выхода передачи |
| Название |  название передачи |
| Жанр |  жанр передачи |
| Начало |  время начала передачи |
| Продолжительность |  продолжительность передачи (мин.) |
| Канал | канал на котором выходит передача |
| Рейтинг | среднее количество человек, смотревших передачу, выраженное в процентах от населения |
| Доля | среднее количество человек, смотревших передачу, выраженное в процентах от всей телеаудитории в данный момент времени |
| Охват | количество человек, смотревших передачу минимум 1 минуту, выраженное в процентах от населения |


>*Эти данные помогут ответить как минимум на два вопроса: Какой контент производить? и Где лучше всего разместить рекламу?*

## **Задание 10**

>*Создайте новый признак `Час`, в котором будет содеражаться информация о часе, в котором началась программа (`Начало`).*
>
>*ПОДСКАЗКА: чтобы достать час, нужно сделать срез `время[:индекс двоеточия]`*

```python
df["Час"] = df["Начало"].str.split(':').str[0]

# OR
df["Час"] = df["Начало"].apply(lambda x: x.split(':')[0])
```
