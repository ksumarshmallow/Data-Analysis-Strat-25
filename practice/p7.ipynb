Данные содержат следующие признаки: 

| Признак | Описание |
|------|----------|
| `rollercoaster_type` | тип американских горок |
| `custom_design` | создан ли аттракцион игроком (1 - да, 0 - нет) |
| `excitement_rating` | уровень экстремальности (Very High - очень высокий, High - высокий, Medium - средний, Low - низкий) |
| `intensity` | числовой показатель мощности и резкости ощущений от поездки |
| `nausea` | числовой показатель, измеряющий, насколько аттракцион вызывает тошноту |
| `max_speed` | пиковая скорость вагонеток |
| `avg_speed` | средняя скорость вагонеток на протяжении всей поездки |
| `ride_time` | полная продолжительность поездки от старта до финиша |
| `ride_length` | общая протяженность всего трека |
| `max_pos_gs` | максимальное значение положительных перегрузок, которые вжимают гостей в кресло |
| `max_neg_gs` | максимальное значение отрицательных перегрузок, которые подбрасывают гостей с сидений |
| `max_lateral_gs` | максимальное значение боковых перегрузок, которые действуют на гостей вбок |
| `total_air_time` | суммарное время в секундах, в течение которого гости испытывают невесомость |
| `drops` | общее количество спусков (падений) на трассе |
| `highest_drop_height` | высота самого высокого спуска на трассе |
| `inversions` | количество переворотов на трассе |
